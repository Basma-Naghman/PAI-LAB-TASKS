```python
import pandas as pd
```

```python
columns = [
    "chapter_Number","chapter_English","chapter_Arabic","Section_number","Section_English","Section_Arabic",
     "haidth_number","English_Hadith","English_Isnad","Englist_Matn","Arabic_Hadith","Arabic_Isnad","Arabic_Matn","Arabic_Comment","English_Grade","Arabic_Grade"
]
```

```python
import glob
path ='/content/LK-Hadith-Corpus'
```

```python
files =sorted(glob.glob(path + "//**//*.csv",recursive=True))
```

```python
files
```

```python

import re
```

```python
def clean_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]','',text)
  text = re.sub(r'\s+','',text)
  return text
```

```python
all_hadith = []
for file in files:
  df = pd.read_csv(file,names=columns , skiprows=1)

  df["Clean_Hadith"] = df["English_Hadith"].astype(str).apply(clean_text)
  all_hadith.extend(df[["chapter_Number","chapter_English","chapter_Arabic","Section_number","Section_English","haidth_number","English_Hadith"]].values.tolist())

```

```python
hadith_df = pd.DataFrame(all_hadith,columns =["chapter_Number","chapter_English","chapter_Arabic","Section_number","Section_English","haidth_number","English_Hadith"])

```
